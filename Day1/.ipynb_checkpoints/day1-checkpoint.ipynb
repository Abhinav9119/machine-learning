<h1>Linear Regression<h/1>

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import os
```

```python
os.chdir(r"C:\Users\dbda\Desktop\MACHINE LEARNING\Datasets")
```

```python
df=pd.read_csv("Housing.csv")
df.info()
```

```python
#changing non numerical columns by using pandas get_dummies function
dum_df=pd.get_dummies(df,drop_first=True)
print(dum_df.columns)
```

```python
x=dum_df.drop("price",axis=1)
y=dum_df["price"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=24)
#train_test_split function split data in random so sum random numbers generated sum are chosen to test and train set 
#random sequence is generated based on perticular number called as random seed/state, number generated by computer also called as sudo random number because it depends on seeds
#train_test_split also generate random seed and according to seed number it generate random split 
#after running everytime results are changes 
#to avoid it fix random_state (it can be any number)
```

```python
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
y_pred=lr.predict(xtest)
print(r2_score(ytest,y_pred))
print(mean_absolute_error(ytest,y_pred))
print(mean_squared_error(ytest,y_pred))
```

```python
df=pd.read_csv("Insure_auto.csv")
df
```

```python
x=df[["Home","Automobile"]]
y=df["Operating_Cost"]
```

<h1>polynomial Regression on Boston.csv dataset<h/1>


```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import os
```

```python
os.chdir(r"C:\Users\dbda\Desktop\MACHINE LEARNING\Datasets")
```

```python
df=pd.read_csv("Boston.csv")
df.info()
```

```python
x=df.drop("medv",axis=1)
y=df["medv"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)  #change it to 1 -linear ,2-quadratic,3-?
xpoly=poly.fit_transform(x)
print(poly.get_feature_names_out())
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xpoly,y,test_size=0.3,random_state=24)
```

```python
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
#degree 1
ypred=lr.predict(xtest)
print(r2_score(ytest,ypred))
```

```python
#degree 2
ypred=lr.predict(xtest)
print(r2_score(ytest,ypred))
```

```python
#degree 3
ypred=lr.predict(xtest)
print(r2_score(ytest,ypred))
```

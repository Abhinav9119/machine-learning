```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import os
```

```python
df=pd.read_csv("Concrete_Data.csv")
df.info()
```

```python
x=df.drop("Strength",axis=1)
y=df["Strength"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)  #change it to 1 -linear ,2-quadratic,3-?
xpoly=poly.fit_transform(x)
print(poly.get_feature_names_out())
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xpoly,y,test_size=0.3,random_state=24)
```

```python
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
#degree 1
ypred=lr.predict(xtest)
print(r2_score(ytest,ypred))
```

```python
#degree 2
ypred=lr.predict(xtest)
print(r2_score(ytest,ypred))
```

```python
#degree 3
ypred=lr.predict(xtest)
print(r2_score(ytest,ypred))   #degree 3 is giving best result
```

```python
#degree 4

ypred=lr.predict(xtest)
print(r2_score(ytest,ypred))
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import os
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("ChemicalProcess.csv")
```

```python
df.isna().sum()
```

```python
x=df.drop("Yield",axis=1)
y=df["Yield"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=24)

```

```python
#imputing 
#train set operations
imp = SimpleImputer(strategy='mean').set_output(transform="pandas")
t_trn_imp = imp.fit_transform(xtrain)  
#fit transform only work on x part not work on y parts
```

```python
#train set operations
lr=LinearRegression()
lr.fit(t_trn_imp,ytrain)
```

```python
#test set operations
x_tst_imp=imp.transform(xtest)
y_pred=lr.predict(x_tst_imp)
print(r2_score(ytest,y_pred))
print(mean_absolute_error(ytest,y_pred))
print(mean_squared_error(ytest,y_pred))
```

# pipeline object:
-it is class
-parameters -1) list of touple



```python
from sklearn.pipeline import Pipeline
```

```python

pipe=Pipeline([("IMP",imp),("LR",lr)])
pipe.fit(xtrain,ytrain)  #train set ops
y_pred=pipe.predict(xtest) #test set ops
print(r2_score(ytest,y_pred))
```

## Pipeline with 3 processes implputer, polynomial ,lr

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
pipe=Pipeline([("IMP",imp),("POLY",poly),("LR",lr)])

pipe.fit(xtrain,ytrain)  #train set ops
y_pred=pipe.predict(xtest) #test set ops
print(r2_score(ytest,y_pred))
```

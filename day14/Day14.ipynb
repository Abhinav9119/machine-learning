# gradiant descend algo without any function

```python
import numpy as np
```

```python
dt=np.array([[0.2,5.6],
[1.2,8.6],
[1,8],
[1.4,9.2],
[-1.5,0.5],
[0.5,6.5],
[-0.5,3.5]])
```

```python
x,y=dt[:,0],dt[:,1]
```

```python
b,w =0.1,0.3
```

```python
y_pred=x*w+b
```

```python
L=np.mean((y-y_pred)**2)/2
eta=0.2
err=y-y_pred
err_x=err*x
```

```python
dw=-(1/len(x)*sum(err_x))
dw
```

```python
db=-(1/len(x))*sum(err)
db
```

```python
new_w=w-eta*dw
new_b=b-eta*db
print(new_w,new_b)
```

```python
b,w =0.1,0.3
eta=0.2
n=len(x)
losses=[]
for i in range(100):
    y_pred=x*w+b
    L=np.mean((y-y_pred)**2)/2
    
    err=y-y_pred
    err_x=err*x
    dw=-(1/len(x)*sum(err_x))
    db=-(1/len(x))*sum(err)
    new_w=w-eta*dw
    new_b=b-eta*db
    b=new_b
    w=new_w
    print(f"w: {new_w:.4f}, b: {new_b:.4f}, L: {L:.4f}")
    print("epoch: ",len(losses)+1)
    losses.append(L)
    if L<0.0001:
        break
    
```

```python
import matplotlib.pyplot as plt
plt.scatter(np.arange(1,len(losses)+1),losses)
plt.plot(np.arange(1,len(losses)+1),losses, color='hotpink')
plt.xlabel("epaches")
plt.ylabel("Loss")
plt.title("Gradient Descent Learning curve")

plt.show()
```

```python
df.head()
```

```python
x1=df.Income
x2=df.Lot_Size
y=df.Response

def sigmoid(z):
    return(1/(1+np.exp(-z)))
```

```python
w1,w2, b = -0.1, 0.3, 0.4
n = len(y)
losses=[]

eta = 0.6
for i in range(1,15001):
    act =x1*w1 + x2*w2 + b
    y_pred = sigmoid(act)
    L = log_loss(y,y_pred)
    db = -(1/n)*np.sum((y-y_pred))
    dw1 = -(1/n)*np.sum((y-y_pred)*x1)
    dw2 = -(1/n)*np.sum((y-y_pred)*x2)
    new_w1 = w1 - eta*dw1
    new_w2 = w2 - eta*dw2
    new_b = b - eta*db
    losses.append(L)
    print('Epoch {0}:'.format(i))
    print('w1={0}, w2={1}, b={2}, L={3}'.format(new_w1, new_w2, new_b, L))
    w1,w2,b=new_w1,new_w2,new_b
    if L<0.001:
        break
```

# gradient desend on Riding Movers with library and pipeline

```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier 
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,log_loss

mowers = pd.read_csv('RidingMowers.csv')
le = LabelEncoder()
mowers['Response'] = le.fit_transform(mowers['Response'])
scaler = MinMaxScaler().set_output(transform='pandas')
df=scaler.fit_transform(mowers)

x=mowers.drop("Response",axis=1)
y=mowers["Response"]
```

```python
# split train test data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
#sgd

sgd=SGDClassifier(random_state=24,loss="log_loss")
pipe=Pipeline([("SCL",scaler),("SGD",sgd)])
pipe.fit(xtrain,ytrain)
```

```python
y_pred=pipe.predict(xtest)
```

```python
print(accuracy_score(ytest,y_pred))
```

```python
y_pred_prob=pipe.predict_proba(xtest)[:,1]
print(log_loss(ytest,y_pred_prob))
```

# stochastic gradient desent classifier with grid search,kfold,pipeline
## finding best parameters

```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier 
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,log_loss
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV


mowers = pd.read_csv('RidingMowers.csv')
le = LabelEncoder()
mowers['Response'] = le.fit_transform(mowers['Response'])
scaler = MinMaxScaler().set_output(transform='pandas')
df=scaler.fit_transform(mowers)

x=mowers.drop("Response",axis=1)
y=mowers["Response"]
```

```python
# split train test data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
sgd=SGDClassifier(random_state=24,loss="log_loss")


params={"SGD__penalty":['l2','l1',None,"elasticnet"],'SGD__eta0':[0.1,0.4,0.7],'SGD__learning_rate':["constant","invscaling","adaptive"]}

pipe=Pipeline([("SCL",scaler),("SGD",sgd)])

kfold=StratifiedKFold(n_splits=5,random_state=24,shuffle=True)
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,verbose=3,scoring="neg_log_loss")

gcv.fit(x,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

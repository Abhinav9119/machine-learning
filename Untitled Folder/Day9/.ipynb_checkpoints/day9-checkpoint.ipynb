# clustering

```python
from scipy.cluster.hierarchy import linkage,dendrogram
import matplotlib .pyplot as plt 
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import os
```

```python
os.chdir(r"C:\Users\dbda\Desktop\MACHINE LEARNING\Datasets")
```

```python
df=pd.read_csv("simp_clust.csv",index_col=0)
```

```python
df.head()
```

```python

plt.scatter (df["X1"],df["X2"])
for i in range(0,df.shape[0]):
    plt.text(x=df["X1"].values[i]
             ,y=df["X2"].values[i],
             s=list(df.index[i]))
plt.show()
```

# hierarchical plot

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(df)
link="average"
mergings=linkage(df_scaled,method=link)
dendrogram(mergings,labels =list(df_scaled.index))
plt.title(link+"linkage")
plt.show()
```

# Agglomerative Clustering 


```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
clust =AgglomerativeClustering(n_clusters=4)
clust.fit(df_scaled)
```

### cluster label
#### same labels are for same cluster

```python
print(clust.labels_)
```

```python
df_clust=df.copy()
df_clust["Clust"]=clust.labels_
df_clust["Clust"]=df_clust["Clust"].astype(str)
```

```python
sns.scatterplot(x=df_clust["X1"],y=df_clust["X2"],hue=df_clust["Clust"])
for i in range(0,df.shape[0]):
    plt.text(df_clust['X1'].values[i],df_clust["X2"].values[i],list(df.index)[i])
plt.show()
```

### calculating silhouette score for cluster =4

```python
print(silhouette_score(df_scaled,clust.labels_))
```

### finding all silhouette coefficients

```python
ks=[2,3,4,5]
scores=[]
for i in ks:
    clust=AgglomerativeClustering(n_clusters=i)
    clust.fit(df_scaled)
    scores.append(silhouette_score(df_scaled,clust.labels_))
scores
```

### find best silhouette score and best number of clusters

```python
i_max=np.argmax(scores)
print("Best no of clusters:",ks[i_max])
print("best score:",scores[i_max])
```

# Kmeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as ptl
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import silhouette_score
```

```python
df=pd.read_csv("nutrient.csv",index_col=0)
```

```python
scaler =StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(df)
```

```python
km=KMeans(n_clusters=4,random_state=24)
km.fit(df_scaled)
print(km.labels_)
```

```python
print(silhouette_score(df_scaled,km.labels_))
```

# iterating throug loop to find best cluster value

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(df)
link="average"
mergings=linkage(df_scaled,method=link)
dendrogram(mergings,labels =list(df_scaled.index))
plt.title(link+"linkage")
plt.xticks(rotation=90)
plt.show()
```

```python
ks=[2,3,4,5]
scores=[]
for i in ks:
    clust=KMeans(n_clusters=i,random_state=24)
    clust.fit(df_scaled)
    scores.append(silhouette_score(df_scaled,clust.labels_))
scores
```

```python
i_max=np.argmax(scores)
print("Best no of clusters:",ks[i_max])
print("best score:",scores[i_max])
```

# mini batch kmeans on milk dataset

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
df=pd.read_csv("milk.csv",index_col=0)
df.head()
```

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(df)
link="average"
mergings=linkage(df_scaled,method=link)
dendrogram(mergings,labels =list(df_scaled.index))
plt.title(link+"linkage")
plt.xticks(rotation=90)
plt.show()
```

```python
ks=[2,3,4,5,6,7,8]
scores=[]
for i in ks:
    clust=MiniBatchKMeans(n_clusters=i,random_state=24)
    clust.fit(df_scaled)
    scores.append(silhouette_score(df_scaled,clust.labels_))
scores
```

```python
i_max=np.argmax(scores)
print("Best no of clusters:",ks[i_max])
print("best score:",scores[i_max])
```

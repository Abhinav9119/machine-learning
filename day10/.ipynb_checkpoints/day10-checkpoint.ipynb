# KMeans clustering on iris flower dataset 
### clustering is also called unsupervised multiclass classification

```python
from scipy.cluster.hierarchy import linkage,dendrogram
import matplotlib .pyplot as plt 
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import os
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
os.chdir(r"C:\Users\dbda\Desktop\MACHINE LEARNING\Datasets")
```

```python
df=pd.read_csv("iris.csv")
```

```python
df.head()
```

```python

df=df.drop("Species",axis=1)
```

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(df)
link="average"
mergings=linkage(df_scaled,method=link)
dendrogram(mergings,labels =list(df_scaled.index))
plt.title(link+"linkage")
plt.show()
```

```python
ks=[2,3,4,5,6,7,8]
scores=[]
for i in ks:
    clust=MiniBatchKMeans(n_clusters=i,random_state=24)
    clust.fit(df_scaled)
    scores.append(silhouette_score(df_scaled,clust.labels_))
scores
```

```python
i_max=np.argmax(scores)
print("Best no of clusters:",ks[i_max])
print("best score:",scores[i_max])
```

#  sum of squares 

```python
from sklearn.datasets import load_wine
# data = load_wine(as_frame=True)  # type of boundle
# data.target[[10, 80, 140]]
# data=data.frame   # converting it into dataframe

df=load_wine(return_X_y=True,as_frame=True) # df is type of touple it return till 
X,y=df[0],df[1]   #0=data  1=target
```

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(X)
```

```python
wss=[]
for i in [2,3,4,5,6,7,8,9,10]:
    km=MiniBatchKMeans(n_clusters=i,random_state=24)
    km.fit(df_scaled)
    wss.append(km.inertia_)
    print(km.inertia_)
```

```python
# inertia gives you an funny figure 
# inertia is variation if we do more grouping inirtia(variation) is also decrease 
# so we need to stop clustering at specefic point
```

```python
#elbow method
import matplotlib.pyplot as plt
ks=[2,3,4,5,6,7,8,9,10]
plt.plot(ks,wss,c="lime")
plt.scatter(ks,wss,c="hotpink")
plt.xlabel("No of clusters")
plt.ylabel("wss")
plt.title("scree plot")
plt.show()
```

```python
# after 3 point there is decrease of inertia is slow down there is an we called it as elbow 
# also after point 8 we can observe one elbow but at point 3 we observe strong change in innertia so point 3 is our best cluster
```

# DBSCAN 

```python
from sklearn.cluster import DBSCAN
```

```python
 x=pd.read_csv("milk.csv",index_col=0)
```

```python
x.head()
```

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(x)
```

```python
dbs=DBSCAN(eps=1,min_samples=3)
dbs.fit(df_scaled)
print(dbs.labels_) #-1 are outliers
```

# delete outlier records

```python
df_clust=df_scaled.copy()
```

```python
df_clust["Clust"]=dbs.labels_
```

```python
df_clust_wo=df_clust[df_clust["Clust"]!=1]  #removing outliers
```

```python
if len(df_clust_wo["Clust"].unique())>=2:
    print(silhouette_score(df_clust_wo.iloc[:,:-1],df_clust_wo.iloc[:,-1]))
```

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(x)
```

```python
scores=[]

#loop
eps=[0.1,0.4,0.7,1,1.5,2,3]
min_pcts=[2,3,4,5]
for i in eps:
    for j in min_pcts:
        dbs=DBSCAN(eps=i,min_samples=j)
        dbs.fit(df_scaled)
        df_clust=df_scaled.copy()
        df_clust["Clust"]=dbs.labels_
        df_clust_wo=df_clust[df_clust["Clust"]!=1]  #removing outliers
        if len(df_clust_wo["Clust"].unique())>=2:
            print(silhouette_score(df_clust_wo.iloc[:,:-1],df_clust_wo.iloc[:,-1]))
            scores.append([i,j,silhouette_score(df_clust_wo.iloc[:,:-1],df_clust_wo.iloc[:,-1])])

        
        
```

```python
scores=pd.DataFrame(scores,columns=["eps","min_pt","score"])
```

```python
scores
```

```python
print("best parameters and score: ")
print(scores.sort_values("score").iloc[-1,:])
```

# PCA -principle componenet analysis

### eigen values are equal to pca components
### it is plain transformation o/p is pca data

```python
from sklearn.decomposition import PCA
```

```python
x=pd.read_csv("Milk.csv",index_col=0)
```

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(x)
```

```python
prcomp=PCA().set_output(transform="pandas")
comps=prcomp.fit_transform(df_scaled)  #principle component data
```

```python
print(comps.shape)
print(comps.var()) #pca values
print(prcomp.explained_variance_) #The amount of variance explained by each of the selected components.
tot_var=np.sum(prcomp.explained_variance_) # sum of amount of variance explained by each of the selected components.(total variance)
prop=np.array(prcomp.explained_variance_)/tot_var
print(prop)
perc=prop*100
print(perc)  # percentage of variation explained by PCi

#cumulative sum of percentage 
print(np.cumsum(perc))  #1st value percentage variation explained by 1st column,  #2st value percentage variation explained by 1st 2 column and so on
```

```python
# var_cov=np.cov(df_scaled.T)
# values,vectors=np.linalg.eig(var_cov)
# print(values)  #eigen values
# print(vectors)  
```

```python
#clustering
```

```python
km=MiniBatchKMeans(n_clusters=3,random_state=24)
km.fit(df_scaled)
print(km.labels_)
```

```python
comps["Clust"]=km.labels_
comps["Clust"]=comps["Clust"].astype(str)
```

```python
sns.scatterplot(data=comps,x="pca0",y="pca1",hue="Clust")
plt.show()
```

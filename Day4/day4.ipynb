```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.metrics import mean_squared_error,accuracy_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import log_loss
from sklearn.pipeline import Pipeline
y=dum_hr["left"]
x=dum_hr.drop("left",axis=1)

dum_hr.head()
print(f"perdcentage: {y.value_counts(normalize=True)*100}")
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,stratify=y,random_state=24)   #stratify the ....?
```

```python
from sklearn.metrics import confusion_matrix
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(xtrain,ytrain)
y_pred=knn.predict(xtest)
y_pred_prob=knn.predict_proba(xtest)
print(confusion_matrix(ytest,y_pred))
print(accuracy_score(ytest,y_pred))
print(classification_report(ytest,y_pred))
print(log_loss(ytest,y_pred_prob))
```

# with standard scalling 

```python
from sklearn.metrics import log_loss
from sklearn.preprocessing import StandardScaler
knn=KNeighborsClassifier(n_neighbors=51)
std_scaler=StandardScaler()
x_trn_scl=std_scaler.fit_transform(xtrain)
knn.fit(x_trn_scl,ytrain)
x_tst_scl=std_scaler.transform(xtest)
y_pred_prob=knn.predict_proba(x_tst_scl)
print(log_loss(ytest,y_pred_prob[:,1]))
```

# with pipeline and with standard scalling 

```python
knn=KNeighborsClassifier(n_neighbors=51)
std_scaler=StandardScaler()
pipe=Pipeline([("SCL",std_scaler),("KNN",knn)])  #1st function need to be transformation functions last will be ml function
pipe.fit(xtrain,ytrain)
y_pred_prob=pipe.predict_proba(xtest)
print(log_loss(ytest,y_pred_prob[:,1]))
```

# pipeline with grid search

```python
params={"KNN__n_neighbors":np.arange(1,40)}
knn=KNeighborsClassifier()
pipe=Pipeline([("SCL",std_scaler),("KNN",knn)])
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=24)
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring="neg_log_loss")
gcv.fit(x,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
x=dum_df.drop("price",axis=1)
y=dum_df["price"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
knn =KNeighborsRegressor()
knn.fit(xtrain,ytrain)
y_pred=knn.predict(xtest)
print(r2_score(ytest,y_pred))
```

```python
mm=MinMaxScaler()
pipe=Pipeline([("SCL",mm),("KNN",knn)])
pipe.fit(xtrain,ytrain)
y_pred=pipe.predict(xtest)
print(r2_score(ytest,y_pred))
```

```python
std=StandardScaler()
pipe=Pipeline([("SCL",std),("KNN",knn)])
pipe.fit(xtrain,ytrain)
y_pred=pipe.predict(xtest)
print(r2_score(ytest,y_pred))
```

```python
mm_scaler=MinMaxScaler()
std_scaler=StandardScaler()
knn =KNeighborsRegressor()
from sklearn.model_selection import KFold

params={"KNN__n_neighbors":np.arange(1,20),"SCL":[None,std_scaler,mm_scaler]}

pipe=Pipeline([("SCL",None),("KNN",knn)])

kfold=KFold(n_splits=5,shuffle=True,random_state=24)

gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,verbose=3)
gcv.fit(x,y)

print(gcv.best_params_)
print(gcv.best_score_)
```

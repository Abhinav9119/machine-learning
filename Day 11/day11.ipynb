# pca analysis 

```python
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
import seaborn as sns
from sklearn.datasets import load_wine

df = load_wine(return_X_y=True, as_frame=True)
X, y = df[0], df[1]

X_train, X_test, y_train, y_test = train_test_split(X, y,
         test_size=0.3,stratify=y, random_state=24)

scaler = StandardScaler().set_output(transform='pandas')
X_trn_scl = scaler.fit_transform(X_train)

prcomp = PCA().set_output(transform='pandas')
X_trn_pca = prcomp.fit_transform(X_trn_scl)

X_trn_pca = X_trn_pca.iloc[:,:6]

lr = LogisticRegression()
lr.fit(X_trn_pca, y_train)

X_tst_scl = scaler.transform(X_test)
X_tst_pca = prcomp.transform(X_tst_scl)

X_tst_pca = X_tst_pca.iloc[:,:6]
y_pred = lr.predict(X_tst_pca)
print(accuracy_score(y_test, y_pred))
```

# pca code with using pipeline

```python
from sklearn.pipeline import Pipeline 
```

```python
scaler =StandardScaler()
prcomp=PCA()
lr =LogisticRegression()
pipe=Pipeline([("SCL",scaler),("PCA",prcomp),("ML",lr)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

# pca analysis 

```python
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
import seaborn as sns
from sklearn.datasets import load_wine

df = load_wine(return_X_y=True, as_frame=True)
X, y = df[0], df[1]

X_train, X_test, y_train, y_test = train_test_split(X, y,
         test_size=0.3,stratify=y, random_state=24)

scaler = StandardScaler().set_output(transform='pandas')
X_trn_scl = scaler.fit_transform(X_train)

prcomp = PCA().set_output(transform='pandas')
X_trn_pca = prcomp.fit_transform(X_trn_scl)

X_trn_pca = X_trn_pca.iloc[:,:6]

lr = LogisticRegression()
lr.fit(X_trn_pca, y_train)

X_tst_scl = scaler.transform(X_test)
X_tst_pca = prcomp.transform(X_tst_scl)

X_tst_pca = X_tst_pca.iloc[:,:6]
y_pred = lr.predict(X_tst_pca)
print(accuracy_score(y_test, y_pred))
```

# pca code with using pipeline

```python
from sklearn.pipeline import Pipeline 
```

```python
scaler =StandardScaler()
prcomp=PCA()
lr =LogisticRegression()
pipe=Pipeline([("SCL",scaler),("PCA",prcomp),("ML",lr)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

# train model using Bankrupcy dataset with PCA

```python
import os
os.chdir(r"C:\Users\dbda\Desktop\MACHINE LEARNING\cases\Bankruptcy")
```

```python
df=pd.read_csv("Bankruptcy.csv",index_col=0)
```

```python
df.head()
```

```python
x=df.drop(["D","YR"],axis=1)
y=df["D"]
```

```python
scaler=StandardScaler().set_output(transform="pandas")
df_scaled=scaler.fit_transform(x)
```

```python
prcomp=PCA().set_output(transform="pandas")
comps=prcomp.fit_transform(df_scaled)  #principle component data
```

```python
print(comps.shape)
print(comps.var()) #pca values
print(prcomp.explained_variance_) #The amount of variance explained by each of the selected components.
tot_var=np.sum(prcomp.explained_variance_) # sum of amount of variance explained by each of the selected components.(total variance)
prop=np.array(prcomp.explained_variance_)/tot_var
print(prop)
perc=prop*100
print(perc)  # percentage of variation explained by PCi

#cumulative sum of percentage 
print(np.cumsum(perc))  #1st value percentage variation explained by 1st column,  #2st value percentage variation explained by 1st 2 column and so on
```

```python
#plot
plt.scatter (np.arange(1,25),np.cumsum(perc))
plt.plot (np.arange(1,25),np.cumsum(perc))
plt.xlabel("components")
plt.ylabel("% cumulative variation")
plt.title("scree plot")
plt.show()
# graph show that more than 80% of data can be represented by 6 pca components
```

```python
from sklearn.metrics import log_loss
X_train, X_test, y_train, y_test = train_test_split(x, y,
         test_size=0.3,stratify=y, random_state=24)


scaler =StandardScaler()
prcomp=PCA(n_components=5)
lr =LogisticRegression()
pipe=Pipeline([("SCL",scaler),("PCA",prcomp),("ML",lr)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
y_pred_prob=pipe.predict_proba(X_test)[:,1]
print(accuracy_score(y_test,y_pred))
print(log_loss(y_test,y_pred_prob))
```

# using pipeline

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV


scaler =StandardScaler()
prcomp=PCA(n_components=5)
lr =LogisticRegression()

#parameters
params={
    "PCA__n_components":[4,5,6,7,8],"ML__penalty":["l2",None],"ML__C":[0.5,1,2]}

pipe=Pipeline([("SCL",scaler),("PCA",prcomp),("ML",lr)])
# pipe.fit(X_train,y_train)

#kfold
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=24)

#grid search
svm_gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,verbose=3,scoring="neg_log_loss")
svm_gcv.fit(x,y)

#printing best parameters
print(svm_gcv.best_params_)
print(svm_gcv.best_score_)
```

# PCA in regression

```python
from sklearn.linear_model import LinearRegression

os.chdir(r"C:\Users\dbda\Desktop\MACHINE LEARNING\Datasets")
```

```python
df=pd.read_csv("Boston.csv")
```

```python
x=df.drop("medv",axis=1)
y=df["medv"]
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV


scaler =StandardScaler()
prcomp=PCA(n_components=5)
lr =LinearRegression()

#parameters
params={
    "PCA__n_components":[4,5,6,7,8]}

pipe=Pipeline([("SCL",scaler),("PCA",prcomp),("ML",lr)])
# pipe.fit(X_train,y_train)

#kfold
kfold=KFold(n_splits=5,shuffle=True,random_state=24)

#grid search
svm_gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,verbose=3,scoring="r2")
svm_gcv.fit(x,y)

#printing best parameters
print(svm_gcv.best_params_)
print(svm_gcv.best_score_)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

```

```python
os.chdir(r"C:\Users\dbda\Desktop\MACHINE LEARNING\Datasets")
```

```python
fp_df=pd.read_csv("Faceplate.csv",index_col=0)
fp_df.head()
```

```python
#create frequ3nt itemsets
fp_df=fp_df.astype(bool)
itemsets =apriori(fp_df,min_support=0.2,use_colnames=True)
```

```python
# and convert into rules 
rules= association_rules(itemsets,metric="confidence",min_threshold=0.6)
```

```python
print(rules[["antecedents","consequents","support","lift"]])
```

```python
rules
```

```python
rules.sort_values(by=["lift"],ascending=False).head(6)
```

```python
rule_df=rules.sort_values(by=["lift","confidence"],ascending=False)
```

```python
print(rule_df[["antecedents","consequents","support","confidence","lift"]])
```

```python
relv_rules=rule_df[rule_df["lift"]>1]
```

```python
relv_rules[["antecedents","consequents","support","confidence","lift"]]
```

# arules on housing dataset

```python
housing=pd.read_csv("Housing.csv")
intervals=[(0,50000),(50000,100000),(100000,150000),(150000,200000)]

#binning
bins=pd.IntervalIndex.from_tuples(intervals)
housing["price_slab"]=pd.cut(housing["price"],bins)
```

```python
housing.head()
```

```python
intervals=[(0,50000),(50000,100000),(100000,150000),(150000,200000)]
bins=pd.IntervalIndex.from_tuples(intervals)
housing["area_slab"]=pd.cut(housing["lotsize"],bins)
housing.head()
```

```python


housing.drop(['price','lotsize'], axis=1, inplace=True)

```

```python
housing = housing.astype(object)
fp_df = pd.get_dummies(housing, prefix_sep='=')

```

```python
# create frequent itemsets
itemsets = apriori(fp_df, min_support=0.2, use_colnames=True)
# and convert into rules
rules = association_rules(itemsets, 
                          metric='confidence', min_threshold=0.6)
rules = rules[['antecedents', 'consequents', 
               'support', 'confidence', 'lift']]
rules = rules[rules['lift']>1]
rules = rules.sort_values(by='lift', ascending=False)
rules
```
